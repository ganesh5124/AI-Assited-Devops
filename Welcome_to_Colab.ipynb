```
# prompt: create dictitory type data of fruits and its calories in 10 entries 0.0 type values

fruits_calories = {
    'apple': 52.0,
    'banana': 89.0,
    'orange': 47.0,
    'grape': 69.0,
    'strawberry': 33.0,
    'blueberry': 57.0,
    'watermelon': 30.0,
    'pineapple': 50.0,
    'mango': 60.0,
    'kiwi': 61.0,
}

fruits_calories
```

```
# Conditional Selection
fruit_calories = pd.Series(fruits_calories, name = "Protein")
fruit_calories
```

```
# logical operators
fruit_calories[(fruit_calories>40) | (fruit_calories<50)]
```

```
# Modifying the series
fruit_calories['apple'] = 100
fruit_calories
```
